```python
import pandas as pd
```

### As we saw previously the data was split into regional and national info

```python
import pycountry as pco
```

```python
iso3_name_set = {country.alpha_3 for country in pco.countries}
iso3_name_set.add('XKX')  
```

```python
df = pd.read_csv('../../data/original_data/population-count/population.csv')
```

```python
df_countries = df[df['Country Code'].isin(iso3_name_set)]
df_countries
```

```python
df[df['Country Name']=='Not classified']
```

## Deleting unecessary columns / empty rows

```python
df_countries.drop(df_countries[df_countries['Country Name']=='Not classified'].index, inplace=True)
```

```python
df_countries.drop(df_countries.columns[[2, 3, 69, 70]], axis=1, inplace=True) ## 69,70 ALL Nans 2, 3 duplicated unecessary data
```

```python
df_countries.reset_index(drop=True)
```

```python
df_groups=df[~df['Country Code'].isin(iso3_name_set)]
df_groups
```

```python
df_groups.drop(df_groups[df_groups['Country Name']=='Not classified'].index, inplace=True)
```

```python
df_groups.drop(df_groups.columns[[2, 3, 69, 70]], axis=1, inplace=True) ## 69,70 ALL Nans 2, 3 duplicated unecessary data
```
