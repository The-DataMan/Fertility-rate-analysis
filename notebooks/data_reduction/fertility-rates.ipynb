# WE WILL WORK ON REDUCING THE DATA TO ONLY THE SECTIONS WE NEED

```python
import pandas as pd
```

### As we saw previously the data was split into regional and national info

```python
import pycountry as pco
```

```python
iso3_name_set = {country.alpha_3 for country in pco.countries}
iso3_name_set.add('XKX')
```

```python
df = pd.read_csv('../../data/original_data/fertility-rates/fertility-rate.csv')
```

```python
df_countries = df[df['Country Code'].isin(iso3_name_set)]
df_countries.head(10)
```

```python
df_groups=df[~df['Country Code'].isin(iso3_name_set)]
df_groups.head(10)
```

```python
df_countries.drop(df_countries.columns[[ 2, 3, 68, 69, 70]], axis=1, inplace=True)
df_groups.drop(df_groups.columns[[ 2, 3, 68, 69, 70]], axis=1, inplace=True)
```

```python
df_avg_n=df_countries['1960'].mean()
df_avg_l=df_countries['2023'].mean()
print(f"The mean global birth rate in 1960 was {df_avg_n:.3} while in 2023 it dropped to {df_avg_l:.3}")
```

```python
df_countries.to_csv("../../data/reduced_data/fertility_rates/fertility-countries.csv", index=False)
```

```python
df_groups.to_csv("../../data/reduced_data/fertility_rates/fertility-groups.csv", index=False)
```
