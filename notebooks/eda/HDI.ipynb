```python
import pandas as pd
```

```python
df = pd.read_csv("../../data/original_data/human-development-index/human-development-index.csv")
```

```python
print(df.info())
```

```python
len(df['Entity'].unique())
```

```python
import pycountry
```

```python
iso3_name_set = {country.alpha_3 for country in pycountry.countries}
```

```python
df_countries=df[df['Code'].isin(iso3_name_set)]
```

```python
df_filtered = df_countries[
    (~(df['Human Development Index'].isna()))  
    & (df['Year'] == 2023)  
]
```

## WE HAVE FILTERED DISTINCT COUNTRY VALUES FOR 2018

```python
df_filtered
```

```python
average_hdi_per_entity =( df_filtered.groupby(
    'World region according to OWID',
    as_index=False)['Human Development Index']
    .mean()
    .rename(columns={'Human Development Index':'Average HDI'})
)
average_hdi_per_entity
```

```python
max_hdi_per_region = df_filtered.groupby('World region according to OWID')['Human Development Index'].transform('max')

df_top = df_filtered[df_filtered['Human Development Index'] == max_hdi_per_region]

df_top
```

```python
min_hdi_per_region = df_filtered.groupby('World region according to OWID')['Human Development Index'].transform('min')

df_bottom = df_filtered[df_filtered['Human Development Index'] == min_hdi_per_region]

df_bottom
```
